```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams["figure.dpi"] = 300
g=9.80665
D=0.01305*2
ref_value=0.354

dframe_mesh_sI_N60_8D_Mo848_00=pd.read_csv('ref-Anjos2014.parameter_00000_template_copy_block/bubblepproc.dat', sep='\s+')
dframe_mesh_sI_N60_8D_Mo848_01=pd.read_csv('ref-Anjos2014.parameter_00001_template_copy_block/bubblepproc.dat', sep='\s+')
dframe_mesh_sI_N60_8D_Mo848_02=pd.read_csv('ref-Anjos2014.parameter_00002_template_copy_block/bubblepproc.dat', sep='\s+')
dframe_mesh_sI_N60_8D_Mo848_03=pd.read_csv('ref-Anjos2014.parameter_00003_template_copy_block/bubblepproc.dat', sep='\s+')


plt.plot(dframe_mesh_sI_N60_8D_Mo848_00["time"]/(D/g)**0.5, dframe_mesh_sI_N60_8D_Mo848_00["magVelocity"]/(g*D)**0.5,"-.",label='interFlow_N64_8D_0.1Tc')
plt.plot(dframe_mesh_sI_N60_8D_Mo848_01["time"]/(D/g)**0.5, dframe_mesh_sI_N60_8D_Mo848_01["magVelocity"]/(g*D)**0.5,"-.",label='interFlow_N96_8D_0.1Tc')
plt.plot(dframe_mesh_sI_N60_8D_Mo848_02["time"]/(D/g)**0.5, dframe_mesh_sI_N60_8D_Mo848_02["magVelocity"]/(g*D)**0.5,"-.",label='interFlow_N128_8D_0.1Tc')
plt.plot(dframe_mesh_sI_N60_8D_Mo848_03["time"]/(D/g)**0.5, dframe_mesh_sI_N60_8D_Mo848_03["magVelocity"]/(g*D)**0.5,"-.",label='interFlow_N160_8D_0.1Tc')

dframe_BLUE_Mo848= pd.read_csv('history.dat',sep='\s+',header=None,skipinitialspace=True)
dframe_extract_Mo848=pd.read_csv('Mo848.csv')
plt.plot(dframe_extract_Mo848['x'],dframe_extract_Mo848['Velocity'],label='Anjos2014')
plt.plot(dframe_BLUE_Mo848.iloc[:,0]/(D/g)**0.5, dframe_BLUE_Mo848.iloc[:,2]/(g*D)**0.5,label='BLUE_N256')
plt.axhline(ref_value, color='blue', linestyle = '-.', label='Hua2007')
plt.axhline(2.47*2.73/(1350*0.0261*(D*g)**0.5), color='black',linestyle = '--', label='exp_Bahga1981')
plt.xlim(left=0,right=6.01)
plt.title("Mo=848, nu_max")
plt.ylabel("Normalized buble rising velocity")
plt.xlabel("Normalized time")
#plt.legend()
plt.legend(bbox_to_anchor=(0, -0.122), loc="upper left",ncol=2)#, bbox_to_anchor=(1, 0.5))
```
